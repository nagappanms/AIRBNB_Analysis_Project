```python
# handling null values

df1.fillna({"Total_bedrooms": df1["Total_bedrooms"].median(),
            "Total_beds": df1["Total_beds"].median(),
            "Security_deposit": df1["Security_deposit"].median(),
            "Cleaning_fee": df1["Cleaning_fee"].median(),
            "Host_response_time":"No data found",
            "Host_response_rate":"No data found",
            "Review_scores": df1["Review_scores"].median(),
            "Review_scores_accuracy": df1["Review_scores_accuracy"].mean(),
            "Review_scores_cleanliness": df1["Review_scores_cleanliness"].mean(),
            "Review_scores_checkin": df1["Review_scores_checkin"].mean(),
            "Review_scores_communication": df1["Review_scores_communication"].mean(),
            "Review_scores_location": df1["Review_scores_location"].mean(),
            "Review_scores_value": df1["Review_scores_value"].mean(),
            "Review_scores_rating": df1["Review_scores_rating"].mean()}, inplace=True)
```

```python
#sum of null values
df1.isna().sum()
```

```python
# CHECKING FOR EMPTY VALUES IN COLUMN
columns_with_empty_strings = df1.eq('').any()
columns_with_empty_strings = columns_with_empty_strings[columns_with_empty_strings].index
print("Columns with empty strings:", columns_with_empty_strings.tolist())
```

```python
# Filling Empty values in Description and House rules columns

df1.replace({"Name": {'': 'Name  not given'},
            "Description": {'':'No Description Found'},
            "House_rules": {'': 'No House rules Found'},
            "Host_location":{'': "No data found"},
            "Host_neighbourhood":{'': "No data found"},
            "suburb":{'':"Unknown"},
            "market":{'':"Unknown"},
            "Amenities": {'': 'No Data Available'}}, inplace=True)
```

```python
df1["is_location_exact"]= df1["is_location_exact"].map({False:"No", True:"Yes"})
```

```python
df1["Id"]=df1["Id"].astype(np.int64)
df1["Extra_people"]= df1["Extra_people"].astype(int)
df1["Total_bedrooms"]= df1["Total_bedrooms"].astype(int)
df1["Total_beds"]= df1["Total_beds"].astype(int)
df1["Price"]= df1["Price"].astype(int)
df1["Security_deposit"]=df1["Security_deposit"].astype(int)
df1["Cleaning_fee"]= df1["Cleaning_fee"].astype(int)
df1["Guests_included"]= df1["Guests_included"].astype(int)
df1["Review_scores"]=df1["Review_scores"].astype(int)
df1["Host_id"]=df1["Host_id"].astype(np.int64)
```
